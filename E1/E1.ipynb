#### importaremos csv y matplotlib usaremos el paquete pyplot, tambien el numpy para calcular en la fabricacion de los graficos

```python
import csv
import matplotlib.pyplot as plt
import numpy as np
```

#### Abriremos el archivo de Data y lo leeremos

```python
fechas=[]
Grupo_de_edades={}
with open ('Data/FallecidosEtario.txt.txt') as F:
    fila = csv.reader(F, delimiter=',')
    bandera=1
    for columnas in fila:
        if bandera==1:
            ##print(columnas)
            bandera=0
            for columna in columnas:
                ##print(columna)
                if '202' in columna:
                    fechas.append(columna)
                    ##print ("si")
                ## else: 
                    ##print("no")
        else:
            Grupo_de_edades[columnas[0]]=1
datos={}
for fecha in fechas:
    datos [fecha]={}
    for Grupo_de_edad in Grupo_de_edades:
        datos[fecha][Grupo_de_edad]=0
```

#### Separaremos los datos por su respectivo orden

```python
with open ('Data/FallecidosEtario.txt.txt') as F:
    fila = csv.reader(F, delimiter=',')
    bandera=1
    
    for columnas in fila:
        if bandera==1:
            bandera=0
        else:
            Grupo_de_edad=columnas.pop(0)
            cont=0
            for dato in columnas:
                ##print (fechas[cont]+ " "+Grupo_de_edad+" "+ " " +dato)
                datos[fechas[cont]][Grupo_de_edad]=datos[fechas[cont]][Grupo_de_edad]+int (dato)
                cont+=1
```

### 1. gráfico de los casos en el tiempo por rango etario.

```python
fig,ax=plt.subplots(figsize=(100,10))
ejeX=np.arange(len(fechas))
ax.set_xticks(ejeX)
ax.set_xticklabels(fechas, rotation=90)
valoresX={}
for Grupo_de_edad in Grupo_de_edades:
    valoresX[Grupo_de_edad]=[]
for fecha in fechas:
    for Grupo_de_edad in Grupo_de_edades:
        valoresX[Grupo_de_edad].append(datos[fecha][Grupo_de_edad])
for Grupo_de_edad in Grupo_de_edades:
    ax.plot(valoresX[Grupo_de_edad],label=Grupo_de_edad)
ax.legend()
ax.set_ylabel("CANTIDAD DE CASOS")
ax.set_xlabel("FECHAS")
ax.set_title("CANTIDAD DE CASOS EN EL TIEMPO POR RANGO ETARIO")
```
